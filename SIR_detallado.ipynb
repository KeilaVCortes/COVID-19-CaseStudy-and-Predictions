```python
#Modelado de la propagación de una enfermedad
import matplotlib.pyplot as plt
import numpy as np

#inicio de variables
ndays = 100
 #reducción de tiempo a fracción
#dt = .1 #paso del tiempo en días
dt = .05 #paso del tiempo en días
ntps = int(ndays/dt)
Beta = (1.0/3.0) # implica una tasa de infección
Gamma = (1.0/14.0) # latencia o tasa de recuperación

S = np.zeros(ntps) #SUSCEPTIBLES (inicio de la pandemia en cero) hasta N días
I = np.zeros(ntps) #INFECTADOS
R = np.zeros(ntps) #RECUPERADOS
t = np.arange(ntps)*dt

#iniciando los Vectores
I[0] = 0.001 #porción inicial de infectados
S[0] = 1.- I[0] #Susceptible inicial
R[0] = 0. #Recuperado inicial


#Empezar el ciclo de los días a contar (i+1)
for i in range(ntps-1):
    S[i+1] = S[i] - Beta*(S[i]*I[i])*dt
    I[i+1] = I[i] + (Beta*S[i]*I[i]- Gamma* I[i])* dt
    R[i+1] = R[i] + (Gamma*I[i])*dt

#visualizando la gráfica
fig = plt.figure(1); fig.clf() #Figura 1; CLF significa figura clara
plt.plot(t,S,'r',lw=3, label='Susceptibles') #S van a estar en 'r'ojo, grosor 3, leyenda
plt.plot(t,I,'g',lw=3, label='Infectados') 
plt.plot(t,R,'b',lw=3, label='Recuperados') 
fig.legend(); plt.xlabel('Días'); plt.ylabel('Fracción de la Población')#Creando el cuadro de acotaciones, y etiquetas en los ejes
```
